```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

# Example 1 A simple cost function

## $$f(x) = x^2 + x + 1$$

```python
def f(x):
    return x**2 + x + 1
```

```python
# Make Data 
x_1 = np.linspace(start=-3, stop=3, num=500)
```

```python
plt.xlim([-3,3])
plt.ylim(0,8)
plt.xlabel("X",fontsize = 16)
plt.ylabel("f(x)",fontsize = 16)

plt.plot(x_1,f(x_1))
plt.show()
```

## Slope & Derivatives

```python
def df(x):
    return 2*x + 1
```

```python
# Plot function and derivative side by side
plt.figure(figsize=[15,5])

#subplot row,column,index["IN ONE row Two Column"]
# this means first chart on the left side index 1
plt.subplot(2,1,1)

plt.xlim([-3,3])
plt.ylim(0,8)
plt.title('Cost function', fontsize=17)
plt.xlabel("X",fontsize = 16)
plt.ylabel("f(x)",fontsize = 16)
# on the left cost_function 
plt.plot(x_1,f(x_1),color='red',linewidth=3)

# Derivative Function
# this means second chart on the left side index 2 derivative function
plt.subplot(2,1,2)
plt.title('Slog of Cost Function', fontsize=17)
plt.xlabel("X",fontsize = 16)
plt.ylabel("df(x)",fontsize = 16)
plt.grid()
plt.xlim(-2,3)
plt.ylim(-3,6)

plt.plot(x_1,df(x_1),color='green',linewidth=3)
plt.show()
```
