# Example 1 A simple cost function

## $$f(x) = x^2 + x + 1$$

```python
def f(x):
    return x**2 + x + 1
```

```python
# Make Data 
x_1 = np.linspace(start=-3, stop=3, num=500)
```

```python
plt.xlim([-3,3])
plt.ylim(0,8)
plt.xlabel("X",fontsize = 16)
plt.ylabel("f(x)",fontsize = 16)

plt.plot(x_1,f(x_1))
plt.show()
```

## Slope & Derivatives

```python
def df(x):
    return 2*x + 1
```

```python
# Plot function and derivative side by side
plt.figure(figsize=[15,5])

#subplot row,column,index["IN ONE row Two Column"]
# this means first chart on the left side index 1
plt.subplot(1,2,1)

plt.xlim([-3,3])
plt.ylim(0,8)
plt.title('Cost function', fontsize=17)
plt.xlabel("X",fontsize = 16)
plt.ylabel("f(x)",fontsize = 16)
# on the left cost_function 
plt.plot(x_1,f(x_1),color='red',linewidth=3)

# Derivative Function
# this means second chart on the left side index 2 derivative function
plt.subplot(1,2,2)
plt.title('Slog of Cost Function', fontsize=17)
plt.xlabel("X",fontsize = 16)
plt.ylabel("df(x)",fontsize = 16)
plt.grid()
plt.xlim(-2,3)
plt.ylim(-3,6)

plt.plot(x_1,df(x_1),color='green',linewidth=3)
plt.show()
```

## Python Loops and Gradient Descent 

```python
for i in range(10):
    print(i* '*',"Keep Going")
count = 10
while count >= 0:
    print(count * '*',"Keep Going")
    count -= 1
```

```python
# Gradient Descent
new_x = 3
previous_x = 0
step_multiplier = .1
precision = 0.00001

x_list = [new_x]
slope_list = [df(new_x)]



for i in range(500):
    previous_x = new_x
    gradient = df(previous_x)
    new_x = previous_x - step_multiplier * gradient
#     print("previous_x = ", previous_x, ">>new_x = ",new_x)

    step_size = abs(new_x - previous_x)
#     print(step_size)

    x_list.append(new_x)
    slope_list.append(df(new_x))
    
    if step_size < precision:
        print("how many Times Run:",i)
        break
        

print("Local minimum in specific range occurs at: ", new_x)
print("Slop or df(x) at this point is: ",df(new_x))
print("f(x) value or cost function at this point: ", f(new_x))
```

```python
# superimpose the gradient descent calculations on plot 

plt.figure(figsize=[20,5])

#subplot row,column,index["IN ONE row Two Column"]
# this means first chart on the left side index 1
plt.subplot(1,3,1)

plt.xlim([-3,3])
plt.ylim(0,8)
plt.title('Cost function', fontsize=17)
plt.xlabel("X",fontsize = 16)
plt.ylabel("f(x)",fontsize = 16)

# on the left cost_function 
plt.plot(x_1,f(x_1),color='red',linewidth=3,alpha=.6)
values = np.array(x_list)
plt.scatter(x_list,f(values), color='black',s=100, alpha= .6)


# Derivative Function
# this means second chart on the left side index 2 derivative function
plt.subplot(1,3,2)
plt.title('Slog of Cost Function', fontsize=17)
plt.xlabel("X",fontsize = 16)
plt.ylabel("df(x)",fontsize = 16)
plt.grid()
plt.xlim(-2,3)
plt.ylim(-3,6)

plt.plot(x_1,df(x_1),color='green',linewidth=3,alpha=.6)
plt.scatter(x_list,slope_list,color='red',s=100,alpha=.5)






#  chart 3
plt.subplot(1,3,3)
plt.title('Gradient Descent ', fontsize=17)
plt.xlabel("X",fontsize = 16)
plt.ylabel("df(x)",fontsize = 16)
plt.grid()
plt.xlim(-0.55,-.2)
plt.ylim(-.3,.8)

plt.plot(x_1,df(x_1),color='green',linewidth=6,alpha=.8)
plt.scatter(x_list,slope_list,color='red',s=300,alpha=.6)



plt.show()
```
