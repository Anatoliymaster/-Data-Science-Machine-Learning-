```python
import pandas
from pandas import DataFrame
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = pandas.read_csv('score_data.csv')
```

```python
time = data[['Time_Delay_in_Minutes']]
LSD = data[['LSD_ppm']]
score = data[['Avg_Math_Test_Score']]
```

```python
%matplotlib inline
plt.title("Tissue concentration of LSD over time",fontsize=17)
plt.xlabel("Time in Minute", fontsize=15,color='#8E44AD')
plt.ylabel("Tissue LSD ppm", fontsize=15,color='#8E44AD')

plt.text(x=0,y=-.5,s="Wagner et al. (1968)",color='#1A5276')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.ylim(1,7) # changerange
# plt.ylim(0,500)

# change the style
plt.style.use('ggplot')
plt.plot(time, LSD, color='#b71c1c', linewidth=2)
```

```python
# a = {'time':[1,2,3,4,5]}
# b = {'speed':[10,5,4,40,20]}
# plt.plot(a['time'],b['speed'])
```

```python
regr = LinearRegression()
regr.fit(LSD,score)
print("Theta1", regr.coef_[0][0])
print("Intercept Y:",regr.intercept_[0])
# how fit 
print("R-Square: ", regr.score(LSD,score))

predicted_score = regr.predict(LSD)
```

```python
%matplotlib inline
plt.title("Arithmetic vs LSD-25",fontsize=17)
plt.xlabel("Tissue LSD ppm", fontsize=15,color='r')
plt.ylabel("Performance Score", fontsize=15,color='r')
plt.ylim(25,85)
plt.xlim(1,7)
plt.style.use('fivethirtyeight')
plt.plot(LSD, predicted_score)
plt.scatter(LSD,score, color='blue', s=100, alpha=.7)
```
